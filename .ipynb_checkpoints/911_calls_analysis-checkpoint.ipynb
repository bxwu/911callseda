# Exploratory Data Analysis of 911 Calls


## Introduction

This is an exploratory data analysis of emergency calls in Montogomery County, PA, which is the third most populous county in the state. The dataset used is hosted on [Kaggle](https://www.kaggle.com/mchirico/montcoalert) and was made avaiable by Mike Chirico. A partial dataset was used it contains the following fields:

* **lat** : String variable, Latitude
* **lng**: String variable, Longitude
* **desc**: String variable, Description of the Emergency Call
* **zip**: String variable, Zipcode
* **title**: String variable, Title
* **timeStamp**: String variable, YYYY-MM-DD HH:MM:SS
* **twp**: String variable, Township
* **addr**: String variable, Address
* **e**: String variable, Dummy variable (always 1)

## Data and Cleanup

Importing the necessary libraries:

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Reading the dataset and checking .info() and .head(). As mentioned in the introduction, the column '**e**' is a dummy variable so it can dropped.

```python
df = pd.read_csv('911.csv').drop(['e'], axis=1)
```

```python
df.info()
```

```python
df.head()
```

Based on the output from .info, we see that there are many null values in the 'zip' column. However, I will proceed without dropping any rows as there may be useful information that can be extracted from the other columns. 

Also noted is the fact that the 'timeStamp' column has a datatype of object.. I'll convert these columns to a datetime datatype which will be useful later.

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df.head()
```

```python
df.info()
```

Everything looks good and we'll proceed by answering some basic questions first!

Top 5 zipcodes with the most 911 calls:

```python
df['zip'].value_counts(ascending=False).head()
```

Top 5 townships with the most 911 calls:

```python
df['twp'].value_counts(ascending=False).head()
```

Number of unique emergency code titles:

```python
df['title'].nunique()
```

## Creating New Features and Visualizing Them

In the 'title' column there is a reason specified before the description - EMS, Fire, or Traffic. I will create a new column called 'reason' to extract those values.

```python
df['reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

```python
df.head()
```

Counts of 911 call reason:

```python
df['reason'].value_counts(ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
plt.title('Total Number of 911 Calls by Reason')
sns.set_palette('muted')
reason_countgraph = sns.countplot(x = 'reason', data = df, order = df['reason'].value_counts().index)
```

Let's also create 3 new columns for month, day, and day of week.

```python
df['month'] = df['timeStamp'].apply(lambda x: x.month)
df['dayofweek'] = df['timeStamp'].apply(lambda x: x.weekday())
df['hour'] = df['timeStamp'].apply(lambda x: x.hour)
```

```python
df.head()
```

We see that the 'dayofweek' column is an integer from 0-6. We can map these integers to the corresponding day of the week:

```python
day_map = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}

# Sorted the dataframe for use in the visualization later
df_sorted = df.sort_values(by=['dayofweek']).copy()
df_sorted['dayofweek'] = df_sorted['dayofweek'].apply(lambda x: day_map[x])

df['dayofweek'] = df['dayofweek'].apply(lambda x: day_map[x])
```

```python
df.head()
```

Looks good, we can now visualize the number of 911 calls by day:

```python
plt.figure(figsize=(14,8))
sns.set_style("whitegrid")
plt.title('911 Calls By Day of the Week')
sns.countplot(x ='dayofweek', data = df_sorted, hue = 'reason', palette = 'coolwarm')
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Visualizing 911 calls by month:

```python
plt.figure(figsize=(14,8))
sns.set_style("whitegrid")
plt.title('911 Calls By Month')
sns.countplot(x ='month', data = df, hue = 'reason', palette='Dark2')
legend = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Notice in the second visualization that a few months are missing. We can use a simple lineplot to fills in the missing values.

```python
df2 = df.groupby('month').count()
```

```python
plot1 = sns.lmplot(x = 'month', y='twp', data=df2.reset_index())
plot1.set(xlim=(0, 13))
plot1.set(ylim=(6000, 14000))
```

I'll now create a new column called 'date' that takes the date of 'timeStamp' column to help visualize the number of 911 calls over time overall and by reason.

```python
df['date'] = df['timeStamp'].apply(lambda x: x.date())
```

```python
plt.figure(figsize=(12,7))
sns.set_style("whitegrid")
plt.title('911 Calls By Day (Overall)')
plt.tight_layout()
line1 = sns.lineplot(x = 'date', y = 'count',
             data = pd.DataFrame(data={'date': df.date.unique(), 'count': df.groupby('date')['twp'].count()}))
```

Traffic Related 911 Calls:

```python
df_traffic = pd.DataFrame(data={'date': df.loc[df['reason'] == 'Traffic']['date'].unique(), 
                                'count': df.loc[df['reason'] == 'Traffic'].groupby('date')['twp'].count()})
```

```python
plt.figure(figsize=(12,7))
sns.set_style("whitegrid")
plt.title('911 Calls By Day (Traffic)')
plt.tight_layout()
line2 = sns.lineplot(x = 'date', y = 'count', data = df_traffic)
```

Fire Related 911 Calls:

```python
df_fire = pd.DataFrame(data={'date': df.loc[df['reason'] == 'Fire']['date'].unique(), 
                                'count': df.loc[df['reason'] == 'Fire'].groupby('date')['twp'].count()})
```

```python
plt.figure(figsize=(12,7))
sns.set_style("whitegrid")
plt.title('911 Calls By Day (Fire)')
plt.tight_layout()
line2 = sns.lineplot(x = 'date', y = 'count', data = df_fire)
```

EMS Related 911 Calls:

```python
df_ems = pd.DataFrame(data={'date': df.loc[df['reason'] == 'EMS']['date'].unique(), 
                                'count': df.loc[df['reason'] == 'EMS'].groupby('date')['twp'].count()})
```

```python
plt.figure(figsize=(12,7))
sns.set_style("whitegrid")
plt.title('911 Calls By Day (EMS)')
plt.tight_layout()
line2 = sns.lineplot(x = 'date', y = 'count', data = df_ems)
```

Finally let's create a couple of heatmap displaying the frequency of the 911 calls at certain hours of the day by month and day of the week.

```python
df_hours = df.groupby(['dayofweek', 'hour'])['twp'].count().unstack(level=-1)
df_hours.head()
```

```python
plt.figure(figsize=(12,7))
heatmap1 = sns.heatmap(data=df_hours, cmap='flare', linewidths=.5)
```

```python
df_months = df.groupby(['dayofweek', 'month'])['twp'].count().unstack(level=-1)
df_months.head()
```

```python
plt.figure(figsize=(12,7))
heatmap2 = sns.heatmap(data=df_months, cmap='flare', linewidths=.5)
```
